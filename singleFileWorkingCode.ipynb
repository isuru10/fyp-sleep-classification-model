```python
import itertools
import datetime
import numpy as np

dates = []
labels = []
allDates = []
readings = []
```

```python
data = np.loadtxt("../Sleep Data/raw_data_1/101/xethru_recording_20171122_231439_x4m200_17FPS_P101_nattbord_100000010083/xethru_sleep_20171122_231440.csv", delimiter=';', unpack = True, skiprows=1, dtype=str)
```

```python
for value in data[0]:
    date = value[11:19]
    dates.append(date)
```

```python
for value in data[3]:
    readings.append(value)
```

```python
with open("../Sleep Data/raw_data_1/101/PSG_analysis_manual/Sleep profile.txt") as g:
    for line in itertools.islice(g, 1, 2):
        print(line)
        e_time = line[23:31]
```

```python
with open("../Sleep Data/raw_data_1/101/PSG_analysis_manual/Sleep profile.txt") as g:
    for line in itertools.islice(g, 8, None):
        new = line[14:-1]
        labels.append(new)
```

```python
endTime_obj = datetime.datetime.strptime(e_time, '%H:%M:%S') + datetime.timedelta(0,30)
end_time = endTime_obj.time()

data = []
allData = []

for i in range(0,len(dates)):
    date_str = dates[i]
    reading = readings[i]
    date_time_obj = datetime.datetime.strptime(date_str, '%H:%M:%S')
    start_time = date_time_obj.time()
    
    print("Start time: ",start_time," End time: ",end_time)
    
    if start_time < end_time:
        data.append(reading)
    else:
        allData.append(data)
        data = []
        data.append(reading)
        endTime_obj = endTime_obj + datetime.timedelta(0, 30)
        end_time = endTime_obj.time()
        print("else: ", end_time)
        
    if i == len(dates)-1:
        allData.append(data)
    else:
        pass
```
